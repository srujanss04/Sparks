Task:Prediction using Supervised ML

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.linear_model import LinearRegression
```

Importing csv file containing data

```python
data =pd.read_csv(r"https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
data.head()
```

```python
data.describe()
```

```python
x = data['Hours']
y = data['Scores']
```

Plotting distribution of scores


```python
plt.scatter(x,y)
plt.xlabel("Hours", fontsize=20)
plt.ylabel("Scores", fontsize=20)
plt.show()
```

```python
x_matrix = x.values.reshape(-1,1)
```

Splitting the data into training and testing sets

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x_matrix, y, test_size=0.2, random_state=0)
```

Training the Algorithm


```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train, y_train)
```

```python
line = regressor.coef_*x+regressor.intercept_
plt.scatter(x,y)
plt.plot(x,line)
plt.show()
```

```python
print(x_test)
y_pred=regressor.predict(x_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

Predicting score of a student who studies 9.25 hours per day

```python
my_pred=regressor.predict(np.array([9.25]).reshape(1,1))
```

```python
print('Predicted score for 9.25 hours per day= {}'.format(my_pred[0]))
```

Evaluating the model

```python
from sklearn import metrics
mean_absolute_error=metrics.mean_absolute_error(y_test,y_pred)
print('mean absolute error =',mean_absolute_error)
```
